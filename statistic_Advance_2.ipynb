```python
#1 What is hypothesis testing in statistics
"""
Hypothesis testing is a way in statistics to decide something about a large group (called a population)
 by looking at a smaller part of it (a sample). It starts with two possible ideas:

-The null hypothesis (H₀), which usually says there's no effect or no difference, and

-The alternative hypothesis (H₁), which says there is an effect or a difference.

Then, we use the sample data to check if we have enough evidence to reject the null
hypothesis and support the alternative one.
"""
```

```python
#2 What is the null hypothesis, and how does it differ from the alternative hypothesis
"""
The null hypothesis (H₀) is a claim that nothing has changed or that there's no real difference or
 effect in the population. It assumes that any changes we see in the data are just due to random chance.

The alternative hypothesis (H₁ or Ha) says the opposite—that there is a real effect or difference.

In hypothesis testing, we use the data to see if there's strong enough evidence to say the null
 hypothesis is likely wrong, and instead accept the alternative one.
"""
```

```python
#3 What is the significance level in hypothesis testing, and why is it important
"""
The significance level, shown as α (alpha), is the chance of making a mistake by rejecting the null
 hypothesis when it's actually true. This kind of mistake is called a Type I error.

A common value for alpha is 0.05, which means there's a 5% risk of wrongly saying there's an effect
 when there isn't one. This level helps us decide what counts as statistically significant, and it plays
  a key role in managing the chances of making either Type I or Type II errors in our conclusions.
"""
```

```python
#4 What does a P-value represent in hypothesis testing
"""
The P-value tells us how likely it is to get results as extreme as the ones we observed, assuming the
 null hypothesis is true. It shows how much evidence we have against the null hypothesis.

A small P-value means it's unlikely our results happened just by chance, so we have stronger reason to
 reject the null hypothesis. Usually, if the P-value is smaller than the significance level (α)—like
  0.05—we reject the null hypothesis.
"""
```

```python
#5  How do you interpret the P-value in hypothesis testing
"""
– The P-value shows how strong the evidence is against the null hypothesis.
– If the P-value is small (usually less than 0.05), it means the results are unlikely if the null hypothesis were true, so we reject H₀.
– If the P-value is large, it means the results are consistent with the null hypothesis, so we do not reject H₀.
"""
```

```python
#6 What are Type 1 and Type 2 errors in hypothesis testing
"""
– A Type 1 error happens when we reject the null hypothesis even though it's actually true.
 This is like a false alarm—the test says there's an effect, but there really isn't one.

– A Type 2 error happens when we don't reject the null hypothesis, but it’s actually false.
 This is like missing something real—the test fails to spot an effect that actually exists.
"""
```

```python
#7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing
"""
– A one-tailed test looks to see if a value is either bigger or smaller than a specific number—it
 only checks in one direction.
– A two-tailed test checks if a value is different from a specific number in either direction, whether
 it's higher or lower. It looks for any significant change, not just one-sided.
"""
```

```python
#8 What is the Z-test, and when is it used in hypothesis testing
"""
The Z-test is a statistical test that helps figure out if there's a significant difference between
 the average of a sample and the average of the population, especially when we know the population's
 variance. It's usually used when the sample size is large (more than 30) or when we already know the
 population's variance.
"""
```

```python
#9 How do you calculate the Z-score, and what does it represent in hypothesis testing
"""
The Z-score is found by taking the difference between the sample mean and the population mean, then dividing
 it by the population standard deviation. It tells us how far the sample mean is from the population mean in
 terms of standard deviations. In hypothesis testing, the Z-score helps us understand if the sample data is
 normal or unusual when assuming the null hypothesis is true.
"""
```

```python
#10 What is the T-distribution, and when should it be used instead of the normal distribution
"""
The T-distribution is a type of probability distribution used when the sample size is small
(less than 30) and we don't know the population variance. It looks like the normal distribution
but has wider tails, which helps to reflect the greater uncertainty that comes with working with
small sample sizes.
"""
```

```python
#11 What is the difference between a Z-test and a T-test
"""
A Z-test is used when the sample size is large and we know the population variance.
On the other hand, a T-test is used when the sample size is small or when the population variance
is unknown. The T-test relies on the T-distribution, which helps to account for the extra uncertainty in small samples.
"""
```

```python
#12 What is the T-test, and how is it used in hypothesis testing
"""
The T-test helps us find out if there's a significant difference between the average values of two groups or
between a sample mean and a population mean. It is typically used when the sample size is small and we don't
know the population standard deviation. There are different types of T-tests, including the one-sample T-test,
independent two-sample T-test, and the paired T-test.
"""
```

```python
#13 What is the relationship between Z-test and T-test in hypothesis testing
"""
Both the Z-test and T-test are used to compare a sample mean to a population mean, but they are used
in different situations. The Z-test is used when the sample size is large or when we know the population
standard deviation. The T-test is used when the sample size is small or when the population standard
deviation is unknown.
"""
```

```python
#14 What is a confidence interval, and how is it used to interpret statistical results
"""
A confidence interval is a range of values that helps us estimate the true value of a population parameter
with a certain level of confidence (like 95% confidence). It gives us a range of possible values for the
parameter and shows how much uncertainty there is in our estimate.
"""
```

```python
#15 What is the margin of error, and how does it affect the confidence interval
"""
The margin of error is how much the sample estimate could differ from the true population value.
It's usually calculated using the standard error and the critical value for the chosen level of confidence.
A larger margin of error means the confidence interval will be wider, which shows more uncertainty in the estimate.
"""
```

```python
#16 How is Bayes' Theorem used in statistics, and what is its significance
"""
Bayes' Theorem is used to adjust the probability of a hypothesis when new evidence becomes available.
It gives us a way to update our beliefs about the likelihood of events based on new information.
This theorem is important because it allows for ongoing learning and updating of statistical
models as more data is gathered.
"""
```

```python
#17 What is the Chi-square distribution, and when is it used
"""
The Chi-square distribution is a type of probability distribution used in hypothesis testing,
especially for tests like independence and goodness-of-fit. It’s used when the data is categorical
and helps us check if the observed frequencies differ significantly from the expected frequencies.
"""
```

```python
#18 What is the Chi-square goodness of fit test, and how is it applied
"""
The Chi-square goodness of fit test helps us check if a sample data set matches an expected probability
distribution. It compares the observed frequencies with the expected frequencies based on a proposed
distribution to see if the data significantly differs from what we expect.
"""
```

```python
#19 What is the F-distribution, and when is it used in hypothesis testing
"""
The F-distribution is used in hypothesis testing to compare the variances of two or more groups.
It is commonly used in ANOVA tests and regression analysis to determine if the variances between
the groups are the same.
"""
```

```python
#20 What is an ANOVA test, and what are its assumptions
"""
ANOVA (Analysis of Variance) is a statistical test used to compare the averages of three or
more groups. It assumes that the data is normally distributed, the observations are independent,
and the variances in the groups are equal.
"""
```

```python
#21 What are the different types of ANOVA tests
"""
The main types of ANOVA tests are:

• One-way ANOVA: Compares the averages of more than two groups based on one factor.
• Two-way ANOVA: Compares the averages based on two factors.
• Repeated measures ANOVA: Compares the averages when the same subjects are tested under
  different conditions.
"""
```

```python
#22 What is the F-test, and how does it relate to hypothesis testing
"""
The F-test is used to compare the variances of two or more groups to see if they likely come
from populations with the same variance. It’s often used in ANOVA and regression analysis to
check whether the model explains a significant amount of variation in the data.
"""
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def compute_standard_scores(values):
    avg = np.mean(values)
    std_dev = np.std(values)
    standard_scores = (values - avg) / std_dev
    return standard_scores

# Generate sample data from a normal distribution
sample_data = np.random.normal(0, 1, 1000)

# Compute Z-scores (standard scores)
standardized_data = compute_standard_scores(sample_data)

# Plot the standardized values
plt.hist(standardized_data, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title('Standard Scores (Z-scores)')
plt.xlabel('Standard Score')
plt.ylabel('Probability Density')
plt.show()
```

```python
#19  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Generate sample data
random_sample = np.random.normal(50, 10, 100)

# Population parameters
pop_mean = 50
pop_std = 10

# Calculate sample statistics
mean_sample = np.mean(random_sample)
std_sample = np.std(random_sample)
size_sample = len(random_sample)

# Z-statistic calculation
z_statistic = (mean_sample - pop_mean) / (std_sample / np.sqrt(size_sample))
print(f"Calculated Z-statistic: {z_statistic}")

# Calculate p-value for two-tailed hypothesis test
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print(f"Calculated P-value: {p_value}")
```

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

# Simulate sample data
sample_set = np.random.normal(loc=50, scale=10, size=100)  # Sample mean = 50, Standard deviation = 10
sample_avg = np.mean(sample_set)
known_population_avg = 55  # Known population mean
known_population_std = 10  # Known population standard deviation
sample_count = len(sample_set)

# Calculate Z-score
z_statistic = (sample_avg - known_population_avg) / (known_population_std / np.sqrt(sample_count))

# Calculate P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#3  Implement a one-sample Z-test using Python to compare the sample mean with the population mean
# Simulate sample data
sample_set = np.random.normal(loc=50, scale=10, size=50)
mean_of_sample = np.mean(sample_set)
mean_of_population = 52
std_of_population = 10
sample_count = len(sample_set)

# Z-test calculation
z_value = (mean_of_sample - mean_of_population) / (std_of_population / np.sqrt(sample_count))
p_result = 2 * (1 - stats.norm.cdf(abs(z_value)))

print(f"Z-value: {z_value}")
print(f"P-result: {p_result}")

# Decision making
if p_result < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Generate test sample and population info
test_sample = np.random.normal(loc=50, scale=10, size=100)
mean_sample = np.mean(test_sample)
mean_population = 52
std_population = 10
n = len(test_sample)

# Compute Z-score and P-value
z_statistic = (mean_sample - mean_population) / (std_population / np.sqrt(n))
p_result = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Plot the decision boundary for hypothesis test
z_range = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_range)

plt.plot(z_range, pdf_values, label='Standard Normal Curve')
plt.fill_between(z_range, 0, pdf_values, where=(z_range <= -1.96) | (z_range >= 1.96),
                 color='red', alpha=0.5, label='Critical Region')
plt.axvline(x=z_statistic, color='black', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')
plt.legend()
plt.title('Z-test Hypothesis Decision Area')
plt.xlabel('Z-value')
plt.ylabel('PDF')
plt.show()

print(f"P-result: {p_result}")
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_error_types(alpha_rate=0.05, beta_rate=0.2):
    # Define means, standard deviation, and sample size
    null_mean = 50
    alt_mean = 55
    std_dev = 10
    n_samples = 100

    # Critical z for Type I error (alpha)
    z_crit = stats.norm.ppf(1 - alpha_rate)

    # Z-score for desired power (1 - beta)
    z_power = stats.norm.ppf(1 - beta_rate)

    # X-axis range and null distribution
    z_values = np.linspace(30, 70, 1000)
    null_pdf = stats.norm.pdf(z_values, null_mean, std_dev / np.sqrt(n_samples))

    # Plot null hypothesis distribution
    plt.plot(z_values, null_pdf, label="H₀ Distribution")

    # Shade Type I Error region (alpha)
    plt.fill_between(z_values, 0, null_pdf, where=(z_values <= z_crit) | (z_values >= -z_crit),
                     color='red', alpha=0.5, label='Type I Error')

    # Shade Type II Error region (beta)
    plt.fill_between(z_values, 0, null_pdf, where=(z_values >= z_power),
                     color='green', alpha=0.5, label='Type II Error')

    plt.axvline(x=z_crit, color='black', linestyle='--', label=f'Z-critical: {z_crit:.2f}')
    plt.legend()
    plt.title('Error Types in Hypothesis Testing')
    plt.xlabel('Sample Statistic')
    plt.ylabel('PDF')
    plt.show()

# Execute the visualization function
visualize_error_types()
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def compute_ci(data_array, conf_level=0.95):
    # Compute statistics from the data
    mean_sample = np.mean(data_array)
    std_sample = np.std(data_array, ddof=1)
    n = len(data_array)

    # Compute margin of error using Z-score
    z_crit = stats.norm.ppf(1 - (1 - conf_level) / 2)
    error_margin = z_crit * (std_sample / np.sqrt(n))

    # Return confidence interval
    ci_bounds = (mean_sample - error_margin, mean_sample + error_margin)
    return ci_bounds

# Create synthetic data
data_input = np.random.normal(loc=50, scale=10, size=100)

# Get confidence interval
interval = compute_ci(data_input)
print(f"Confidence Interval: {interval}")
```

```python
#16 Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

# Simulate two random samples from normal distributions
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1: mean=50, std=10, n=100
sample2 = np.random.normal(loc=50, scale=15, size=100)  # Sample 2: mean=50, std=15, n=100

# Calculate the F-statistic (ratio of variances)
var_sample1 = np.var(sample1, ddof=1)
var_sample2 = np.var(sample2, ddof=1)

f_stat = var_sample1 / var_sample2  # F-statistic (ratio of sample variances)
dof1 = len(sample1) - 1  # Degrees of freedom for sample 1
dof2 = len(sample2) - 1  # Degrees of freedom for sample 2

# Calculate the p-value for the F-distribution
p_value = 1 - stats.f.cdf(f_stat, dof1, dof2)

# Output results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import scipy.stats as stats
import numpy as np

def validate_anova_assumptions(*samples):
    # Normality check (Shapiro-Wilk Test)
    for i, sample in enumerate(samples):
        statistic, p_value = stats.shapiro(sample)
        print(f"Sample {i+1} - Normality Test (Shapiro-Wilk): p-value = {p_value}")
        if p_value < 0.05:
            print(f"Sample {i+1}: Not Normally Distributed")
        else:
            print(f"Sample {i+1}: Normally Distributed")

    # Homogeneity of variances (Levene's Test)
    statistic, p_value = stats.levene(*samples)
    print(f"Levene's Test for Equal Variance: p-value = {p_value}")
    if p_value < 0.05:
        print("Warning: The variances are not equal!")
    else:
        print("The variances are equal.")

# Simulate data for 3 different samples
sample_a = np.random.normal(loc=50, scale=10, size=30)
sample_b = np.random.normal(loc=55, scale=10, size=30)
sample_c = np.random.normal(loc=60, scale=10, size=30)

# Validate assumptions
validate_anova_assumptions(sample_a, sample_b, sample_c)
```
